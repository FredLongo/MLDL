```python
import os

# Specify the directory you want to count the files in
directory1 = 'drive/MyDrive/Databases/Dataset_Project_1/train/Class 1 Touch'
directory2 = 'drive/MyDrive/Databases/Dataset_Project_1/train/Class 2 No hands'
directory3 = 'drive/MyDrive/Databases/Dataset_Project_1/train/Class 3 No Touch w Hands'


# List all entries in the directory
entries1 = os.listdir(directory1)
entries2 = os.listdir(directory2)
entries3 = os.listdir(directory3)
# Filter out only the files
files1 = [entry for entry in entries1 if os.path.isfile(os.path.join(directory1, entry))]
files2 = [entry for entry in entries2 if os.path.isfile(os.path.join(directory2, entry))]
files3 = [entry for entry in entries3 if os.path.isfile(os.path.join(directory3, entry))]
# Count the files
file_count1 = len(files1)
file_count2 = len(files2)
file_count3 = len(files3)

print(f"{directory1} {file_count1} * .2 = {file_count1*.2}")
print(f"{directory2} {file_count2} * .2 = {file_count2*.2}")
print(f"{directory3} {file_count3} * .2 = {file_count3*.2}")

```

```python
import os
import shutil
from random import sample

# Define the source and destination directories
source_dir = 'drive/MyDrive/Databases/Dataset_Project_1/train'
validate_dir = 'drive/MyDrive/Databases/Dataset_Project_1/validate'
classes = ['Class 1 Touch' , 'Class 2 No hands' , 'Class 3 No Touch w Hands']

# Ensure the validate directory exists for each class
for class_name in classes:
    class_validate_dir = os.path.join(validate_dir, class_name)
    os.makedirs(class_validate_dir, exist_ok=True)

# Loop through each class directory in the train directory
for class_name in classes:
    print("Starting: ", class_name)
    goal_count = 3650
    current_count = 0
    class_source_dir = os.path.join(source_dir, class_name)
    class_validate_dir = os.path.join(validate_dir, class_name)

    # Get a list of all files in the current class directory
    all_files = [f for f in os.listdir(class_source_dir) if os.path.isfile(os.path.join(class_source_dir, f))]

    # Randomly select 3650 files (or less if there are not enough files)
    selected_files = sample(all_files, min(3650, len(all_files)))

    # Move the selected files to the corresponding class directory in the validate directory
    for filename in selected_files:
        current_count += 1
        src_path = os.path.join(class_source_dir, filename)
        dest_path = os.path.join(class_validate_dir, filename)
        shutil.move(src_path, dest_path)
        print("Moved: ", str(current_count))

print("Files successfully moved to the validate directory.")


```
